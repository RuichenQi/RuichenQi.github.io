```python
import numpy as np
import matplotlib.pyplot as plt
import torch

x = torch.tensor(np.linspace(start=-10, stop=10, num = 200))
Sigmoid = torch.sigmoid(x)

plt.plot(x, Sigmoid)
plt.scatter([0],[0.5])
plt.annotate("P_1 (0, 0.5)",xy=(0,0.5), xytext=(0.6,0.5))

plt.legend(labels=["Sigmoid(x)"])
plt.xlabel("x")
plt.savefig('C:/Users/14244/Documents/GitHub/RuichenQi.github.io/assets/images/AI/1/sigmoid.jpg')
plt.show()
```
